# Vilualisation


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.metrics.pairwise import cosine_similarity
from tensorflow.keras.models import load_model
from itertools import product
```

```python
def calculate_diversity(df):
    """
    Calculate diversity for restaurants.

    Args:
    - df: DataFrame containing business features.

    Returns:
    - Diversity measure (float) calculated based on cosine similarity between business features.
    """
    # Stack business features into a numpy array
    features = np.stack(df['business_features'].values)

    # Calculate cosine similarity matrix between business features
    similarity_matrix = cosine_similarity(features)

    # Extract upper triangular part of the similarity matrix
    upper_tri = similarity_matrix[np.triu_indices_from(similarity_matrix, k=1)]

    # Calculate diversity as 1 minus the mean of upper triangular values
    return 1 - np.mean(upper_tri)
```

## Hawaii Data

```python
# Create a figure with specified size
fig = plt.figure(figsize=(20, 8))
spec = gridspec.GridSpec(ncols=3, nrows=2, figure=fig)

# Load test and train datasets
test_datasets = {
    '10': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/test_final_10.parquet'),
    '20': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/test_final_20.parquet'),
    '50': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/test_final_50.parquet'),
    '80': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/test_final_80.parquet'),
    '100': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/test_final_100.parquet')
}

train_datasets = {
    '10': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/train_final_10.parquet'),
    '20': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/train_final_20.parquet'),
    '50': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/train_final_50.parquet'),
    '80': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/train_final_80.parquet'),
    '100': pd.read_parquet('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/train_final_100.parquet')
}

for i, (percentage, test_data) in enumerate(test_datasets.items()):
    ax = fig.add_subplot(spec[i // 3, i % 3])
    model = load_model('drive/MyDrive/lse/Course Materials/ST446 Distribued Computing/project/hi_data/final data/final/model_{}.h5'.format(percentage))

    train_data = train_datasets[percentage]
    rank = 60

    # Fill None values with zero vectors
    test_data['user_features'] = test_data['user_features'].apply(lambda x: np.zeros(rank) if x is None else x)
    test_data['business_features'] = test_data['business_features'].apply(lambda x: np.zeros(rank) if x is None else x)
    test_data['features_dot'] = test_data.apply(lambda row: np.dot(row['user_features'], row['business_features']), axis=1)
    test_data['dot_is_zero'] = (test_data['features_dot'] == 0).astype(int)
    test_data['user_is_zero'] = test_data['user_features'].apply(lambda features: int(np.array_equal(features, np.zeros(rank))))
    test_data['business_is_zero'] = test_data['business_features'].apply(lambda features: int(np.array_equal(features, np.zeros(rank))))

    # Calculate diversity for user cold starts
    user_cold_start_ids = test_data.loc[~test_data['user_id'].isin(train_data['user_id']), 'user_id'].unique()
    cold_start_data = test_data[test_data['user_id'].isin(user_cold_start_ids)]
    cold_features = [np.array([x for x in cold_start_data['user_features']]), np.array([x for x in cold_start_data['business_features']])]
    y_pred_cold = model.predict(cold_features + [cold_start_data['user_id'], cold_start_data['business_id'],
                                                 cold_start_data['features_dot'], cold_start_data['avg_rating'],
                                                 cold_start_data['num_of_reviews'], cold_start_data['dot_is_zero'],
                                                 cold_start_data['user_is_zero'], cold_start_data['business_is_zero']])
    cold_start_data['predicted_rating'] = y_pred_cold.flatten()

    # Calculate the number of users representing 1% of the total
    one_percent_users = int(len(user_cold_start_ids) * 0.01)

    # Randomly select approximately 1% of unique user_ids
    selected_user_ids = cold_start_data.drop_duplicates(subset=['user_id'])['user_id'].sample(n=one_percent_users, random_state=42).tolist()

    # Extract unique user and business data from the original data
    user_data = cold_start_data[['user_id', 'user_features']].drop_duplicates(subset=['user_id'])
    business_data = cold_start_data[['business_id', 'business', 'avg_rating', 'num_of_reviews', 'business_features']].drop_duplicates(subset=['business_id'])

    # Create all possible combinations of users and businesses
    combined_data = pd.DataFrame(list(product(selected_user_ids, business_data['business_id'])), columns=['user_id', 'business_id'])
    combined_data = combined_data.merge(user_data, on='user_id', how='left')
    combined_data = combined_data.merge(business_data, on='business_id', how='left')

    # Handle missing feature vectors
    rank = 60
    combined_data['user_features'] = combined_data['user_features'].apply(lambda x: np.zeros(rank) if x is None else x)
    combined_data['business_features'] = combined_data['business_features'].apply(lambda x: np.zeros(rank) if x is None else x)

    # Calculate dot product of user and business features
    combined_data['features_dot'] = combined_data.apply(lambda row: np.dot(row['user_features'], row['business_features']), axis=1)

    # Convert dot product to binary indicator if it's zero or not
    combined_data['dot_is_zero'] = (combined_data['features_dot'] == 0).astype(int)

    # Check if user features are all zeros
    combined_data['user_is_zero'] = combined_data['user_features'].apply(lambda features: int(np.array_equal(features, np.zeros(rank))))

    # Check if business features are all zeros
    combined_data['business_is_zero'] = combined_data['business_features'].apply(lambda features: int(np.array_equal(features, np.zeros(rank))))

    # Prepare features for model prediction
    features = [np.array([x for x in combined_data['user_features']]), np.array([x for x in combined_data['business_features']])]

    # Predict ratings using the model
    y_pred = model.predict(features + [combined_data['user_id'], combined_data['business_id'],
                                      combined_data['features_dot'], combined_data['avg_rating'], combined_data['num_of_reviews'],
                                      combined_data['dot_is_zero'], combined_data['user_is_zero'], combined_data['business_is_zero']])

    # Assign predicted ratings to the combined data
    combined_data['predicted_rating'] = y_pred.flatten()

    # Select top 10 recommendations for each user
    top_10_recs = combined_data.groupby('user_id').apply(lambda x: x.nlargest(10, 'predicted_rating')).reset_index(drop=True)

    # Calculate diversity for each user's top 10 recommendations
    user_diversity = top_10_recs.groupby('user_id').apply(calculate_diversity)

    # Calculate the mean diversity across all users for user cold starts
    user_cold_start_diversity = user_diversity.mean()

    # Calculations for non-cold start diversity
    non_cold_start_ids = test_data.loc[test_data['user_id'].isin(train_data['user_id']), 'user_id'].unique()
    non_cold_start_data = test_data[test_data['user_id'].isin(non_cold_start_ids)]
    non_cold_features = [np.array([x for x in non_cold_start_data['user_features']]), np.array([x for x in non_cold_start_data['business_features']])]
    y_pred_non_cold = model.predict(non_cold_features + [non_cold_start_data['user_id'], non_cold_start_data['business_id'],
                                                         non_cold_start_data['features_dot'], non_cold_start_data['avg_rating'],
                                                         non_cold_start_data['num_of_reviews'], non_cold_start_data['dot_is_zero'],
                                                         non_cold_start_data['user_is_zero'], non_cold_start_data['business_is_zero']])
    non_cold_start_data['predicted_rating'] = y_pred_non_cold.flatten()

    # User Cold Start
    precision_cold, recall_cold, _ = precision_recall_curve(cold_start_data['rating_binary'], y_pred_cold)
    ap_cold = average_precision_score(cold_start_data['rating_binary'], y_pred_cold)
    ax.plot(recall_cold, precision_cold, label=f'User Cold Start (Average Precision={ap_cold:.2f}, Diversity={user_cold_start_diversity:.2f}, n={len(user_cold_start_ids)})')

    # Non-Cold Start
    precision_non_cold, recall_non_cold, _ = precision_recall_curve(non_cold_start_data['rating_binary'], y_pred_non_cold)
    ap_non_cold = average_precision_score(non_cold_start_data['rating_binary'], y_pred_non_cold)
    ax.plot(recall_non_cold, precision_non_cold, label=f'Non-Cold Start (Average Precision={ap_non_cold:.2f}, Diversity={non_user_cold_start_diversity:.2f}, n={len(non_cold_start_ids)})')

    # Set the y-axis to range from 0 to 1
    ax.set_ylim(0, 1)

    # Set title with the percentage of data
    ax.set_title(f'{percentage}% Data')
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')

    # Place legend at the bottom left of the plot
    ax.legend(loc='lower left')

plt.tight_layout()
plt.show()
```
